Vamos dividir os gêneros em Dummies para facilitar o processamento futuro

```python
dummies = movies['genres'].str.get_dummies(sep='|')
movies.drop(columns=['genres'], inplace=True)
#print(dummies.head())

movies = pd.concat([movies, dummies], axis=1)
descreve_df(movies)
```

Analisando o cabeçalho, vemos que existe a coluna '(no genres listed)'. Vamos analisá-la um pouco:

Para continuar a análise, vamos juntar os dois DataFrames:

```python
ratings = pd.merge(ratings, movies, how='left', left_on ='movieId' ,right_on = 'movieId')
display(ratings.head())
```

Agora vamos criar uma matriz esparsa de usuários x filmes com as notas como valores e depois converter em objeto Numpy para melhorar o processamento, visto que o Pandas foi construído usando Numpy

```python
rp = ratings.pivot_table(columns='movieId', index='userId', values='rating')
rp = rp.fillna(0)
rp_mat = rp.values

rp_mat.shape
```

```python
# Encontrar usuários similares 
def top_similarusers(df, uid=16,n=5): 
    users = df.loc[uid, :].sort_values(ascending = False) 
    topn_users = users.iloc[:n, ] 
    topn_users = topn_users.rename('Score')     
    print ("Similar users as user:", uid) 
    print(pd.DataFrame(topn_users))
    print('\n')

top_similarusers(df=pd_users, uid=270, n=5)
top_similarusers(df=pd_users, uid=150, n=5)

#top_similarusers(df=pd_users, uid=17, n=5)
```

Vamos encontrar agora os filmes mais bem votados por cada usuário:

```python
#Encontrar melhores filmes do usuário
def topn_movieratings(df, uid, n_ratings=10):
    uid_ratings = df.loc[df['userId']==uid][['userId','rating','title']]
    uid_ratings = uid_ratings.sort_values(by='rating', ascending=False)
    print("Top",n_ratings,"filmes do usuário",uid)
    print(uid_ratings.iloc[:n_ratings,])
    print("\n")
    
topn_movieratings(df=ratings, uid=597, n_ratings=5)
```

<h2>Filtro Colaborativo usando ALS</h2>

```python
A = rp.values
print("Shape of Original Sparse Matrix", A.shape)
```

Vamos agora criar uma matriz W com valor 0 ou 1 sendo que 1 sigfnifica que o usuário deu uma nota ao filme.

```python
W = A > 0.5
W[W==True]=1
W[W==False]=0
W = W.astype(np.float64, copy=False)
print(W)
```

De forma similar, vamos criar uma matriz W_pred com os valores inversos para podermos gerar as sugestões de filmes (nós queremos só sugerir filmes que ele NÃO viu)

```python
W_pred = A < 0.5
W_pred[W_pred == True] = 1
W_pred[W_pred == False] = 0
W_pred = W_pred.astype(np.float64, copy=False)
np.fill_diagonal(W_pred, val=0)
```
